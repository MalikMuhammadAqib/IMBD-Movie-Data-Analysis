Classify Movies Based On Ratings [Excellent,Good And Average]

```python
movie_data.columns
```

```python
def rating(rating):
    if rating >= 7.5 :
         return "Excellent"
    elif rating >= 6.0:
        return "Good"
    else :
        return "Average"
```

```python
movie_data['rating_cat']= movie_data['Rating'].apply(rating)
```

Find Unique Values From Genre

```python
movie_data.columns
```

```python
movie_data['Genre']
```

```python
list1=[]
for value in movie_data['Genre']:
      list1.append(value.split(','))
```

```python
list1
```

```python
one_d=[]
for item in list1:
   for item1 in item:
      one_d.append(item1)
```

```python
one_d
```

```python
uni_list=[]
for item in one_d:
  if item not in uni_list:
    uni_list.append(item)
```
